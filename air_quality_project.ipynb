#### For each city, plot the monthly trend of AQI over 3 years. Highlight the worst and best month per city.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Ensure 'Date' is datetime format
df_filtered['Date'] = pd.to_datetime(df_filtered['Date'])

# Step 2: Create 'YearMonth' column
df_filtered['YearMonth'] = df_filtered['Date'].dt.to_period('M').astype(str)

# Step 3: Calculate monthly average AQI per city
monthly_aqi = df_filtered.groupby(['City', 'YearMonth'])['AQI'].mean().reset_index()

# Step 4: Identify best and worst months per city
highlight_points = []

for city in monthly_aqi['City'].unique():
    city_data = monthly_aqi[monthly_aqi['City'] == city]
    
    # Find best (min) and worst (max) month
    best = city_data.loc[city_data['AQI'].idxmin()]
    worst = city_data.loc[city_data['AQI'].idxmax()]
    
    highlight_points.append((city, best['YearMonth'], best['AQI'], 'best'))
    highlight_points.append((city, worst['YearMonth'], worst['AQI'], 'worst'))

highlight_df = pd.DataFrame(highlight_points, columns=['City', 'YearMonth', 'AQI', 'Type'])

# Step 5: Plot
cities = monthly_aqi['City'].unique()
for city in cities:
    plt.figure(figsize=(10, 5))
    
    city_data = monthly_aqi[monthly_aqi['City'] == city]
    plt.plot(city_data['YearMonth'], city_data['AQI'], marker='o', label='Monthly AQI', color='blue')
    
    # Highlight best and worst months
    city_highlights = highlight_df[highlight_df['City'] == city]
    for _, row in city_highlights.iterrows():
        color = 'green' if row['Type'] == 'best' else 'red'
        label = 'Best Month' if row['Type'] == 'best' else 'Worst Month'
        plt.scatter(row['YearMonth'], row['AQI'], color=color, label=label, zorder=5)
    
    plt.title(f'Monthly AQI Trend for {city}')
    plt.xlabel('Month')
    plt.ylabel('Average AQI')
    plt.xticks(rotation=90)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

### Build a regression model to predict AQI using pollutants (PM2.5, PM10, NOx, CO, etc.)

#### Split data into train-test (e.g., 80-20)


```python
import pandas as pd
data=pd.read_csv('cleaned_data_month_year.csv')
print(data.head())
```

```python
data = data.dropna()
```

```python
print(data.head())
```

```python
from sklearn.model_selection import train_test_split
X = data[['PM2.5', 'PM10', 'NOx', 'CO', 'SO2', 'O3']]
target=data['AQI']
X_train, X_test, y_train, y_test = train_test_split( X, target, test_size=0.2, random_state=42)

print('xtrain',X_train)
print('X_test',X_test)
print('y_train',y_train)
print('X_test',X_test)
```

```python
Import the train_test_split function from the sklearn.model_selection module

X: This is your feature matrix. It includes the columns that will be used to predict AQI
target: This is the target variable or the value you're trying to predict 

X_train, X_test, y_train, y_test = train_test_split( X, target, test_size=0.2, random_state=42)
This splits both X and target into training and testing sets.
```

### Use Linear Regression and Random Forest. Increase accuracy by tuning

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

model = LinearRegression()
model.fit(X_train, y_train)

# Predict and evaluate
predictions = model.predict(X_test)
print('predictions is this',predictions)
print("Linear Regression R2:", r2_score(y_test, predictions))
```
